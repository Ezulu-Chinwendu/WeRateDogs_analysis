## Introduction
### This report shows some insights and visuals from the data wrangling  and analysis of the WeRateDogs twitter account. 

### Insight 1

### Source of Tweet
#### The first insight of this analysis was to see the count of different sources where the tweeet was generated from.
#### During our analysis, we observed that the majority of the tweets originated from the Twitter for iPhone platform, with a count of 2221. Additionally, we identified 91 tweets from Vine, 33 from Twitter Web Client, and 11 from Tweetdeck. These findings provide insights into the distribution of tweet sources within the WeRateDogs Twitter account..


### Insight 2

### Dog Stages According to Favorite Count(Likes)

#### In our analysis, we discovered that the puppo stage received the highest number of likes, totaling 132,810. Following closely behind, the doggos stage garnered 131,075 likes. These findings suggest that the puppo and doggos stages were particularly popular among the audience, as evidenced by the significant engagement in the form of likes.

### Insight 3

### Prediction of Dog stages according to prediction 1, 2 and 3

#### I wanted to see how true the dog predictions were considering they had three different predictions. In prediction 1, the doggo dog stage had 51 True predictions and 16 False predictions. When compared to  prediction 2(52 True and 15 False), and prediction 3(47 True and 20 False), we can see that all the predictions are in the same range. The same thing goes for other dog stages.
#### Based on our findings, it appears that Prediction 1 alone could have provided sufficient information to determine how easily users could identify dog stages through pictures. The strong correlation between the number of likes and specific dog stages suggests that users were able to discern and appreciate the different stages of dogs based on the visual content alone. Therefore, the additional predictions 2 and 3 might not have been necessary in this particular analysis.

### Visual 1

### Correlation between Favorite Count(Likes) and Retweet Count

### Based on the visual analysis, we can observe a positive correlation between the number of likes and retweets on the dog pictures. The data indicates that as the number of likes increases, so does the number of retweets. This suggests that pictures with lower likes are less likely to be retweeted. The positive correlation implies that higher engagement, as reflected by the likes, tends to drive more retweets, indicating the perceived quality or popularity of the dog pictures.
